```python
#convert scentences to words
import re
import nltk
test_text_without_symbols = re.sub('[:\',\-!;"()?âââ|%/]',"", test_text)
test_tokens = nltk.word_tokenize(test_text_without_symbols)
test_tokens = list(map(lambda x: x.lower(), test_tokens))

test_tagged = nltk.pos_tag(test_tokens, tagset='universal')

meaning_collector = []
for i in test_tagged:
    if i[1] == 'NOUN':
        meaning_collector.append((i[0],'n'))
    elif i[1] == 'VERB':
        meaning_collector.append((i[0],'v'))
    elif i[1] == 'ADJ':
        meaning_collector.append((i[0],'a'))
    elif i[1] == 'ADV' :
        meaning_collector.append((i[0], 'r'))
    else:
        pass

nouns = []
verbs = []
adjectives = []
adverbs = []
for i in meaning_collector:
    if i[1] == 'n':
        nouns.append(i)
    elif i[1] == 'v':
        verbs.append(i)
    elif i[1] == 'a':
        adjectives.append(i)
    elif i[1] == 'r':
        adverbs.append(i)
    else:
        pass

lmtz = nltk.stem.WordNetLemmatizer()
lmtzd_nouns = list(map(lambda x: lmtz.lemmatize(x[0], str(x[1])), nouns))
lmtzd_verbs = list(map(lambda x: lmtz.lemmatize(x[0], str(x[1])), verbs))
lmtzd_adjectives = list(map(lambda x: lmtz.lemmatize(x[0], str(x[1])), adjectives))
lmtzd_adverbs = list(map(lambda x: lmtz.lemmatize(x[0], str(x[1])), adverbs))

#stm = nltk.stem.PorterStemmer()
#[stemmer.stem(word) for word in meaning_collector3]
```

```python
finding_word = 'old'
finding_neighbors = []

from more_itertools import locate
word_locations = list(locate((test_tokens), lambda x: x == finding_word))

for i in word_locations:
    period_locations = list(locate((test_tokens), lambda x: x == '.'))
    period_locations.append(i)
    period_locations.sort()
    subs_num = period_locations.index(i)
    finding_neighbors.append(test_tagged[period_locations[subs_num-1]:period_locations[subs_num+1]+1])

finding_neighbors
```

```python
import nltk
lemma = nltk.stem.WordNetLemmatizer()
stemmer = nltk.stem.PorterStemmer()

flatten = sum(finding_neighbors, [])
nested_words = []

for i in flatten:
    if i[1] == 'ADJ':
        lemma_word = lemma.lemmatize(i[0], 'a')
        stem_word = stemmer.stem(lemma_word)
        nested_words.append((i[0],lemma_word,stem_word,'a'))
    if i[1] == 'ADV':
        lemma_word = lemma.lemmatize(i[0], 'r')
        stem_word = stemmer.stem(lemma_word)
        nested_words.append((i[0],lemma_word,stem_word,'r'))
    if i[1] == 'NOUN':
        lemma_word = lemma.lemmatize(i[0], 'n')
        stem_word = stemmer.stem(lemma_word)
        nested_words.append((i[0],lemma_word,stem_word,'n'))
    if i[1] == 'VERB':
        lemma_word = lemma.lemmatize(i[0], 'v')
        stem_word = stemmer.stem(lemma_word)
        nested_words.append((i[0],lemma_word,stem_word,'v'))
    else:
        pass
    
nested_words  #in tuple : (original, lemmatized, stemmized, word class)
```

```python
#finding major neighbors each word class
for_major_adjectives = []
for_major_adverbs = []
for_major_nouns = []
for_major_verbs = []

for i in nested_words:
    if i[3] == 'a':
        for_major_adjectives.append(i[2])
    if i[3] == 'r':
        for_major_adverbs.append(i[2])
    if i[3] == 'n':
        for_major_nouns.append(i[2])
    if i[3] == 'v':
        for_major_verbs.append(i[2])
    else:
        pass

for_major_adjectives_unique = list(set(for_major_adjectives))
arranged_major_adjectives = []
for i in range(0,len(for_major_adjectives_unique)-1):
    a = for_major_adjectives_unique[i]
    arranged_major_adjectives.append((a, for_major_adjectives.count(a)))
arranged_major_adjectives.sort(key=lambda x: x[1], reverse=True)

for_major_adverbs_unique = list(set(for_major_adverbs))
arranged_major_adverbs = []
for i in range(0,len(for_major_adverbs_unique)-1):
    a = for_major_adverbs_unique[i]
    arranged_major_adverbs.append((a, for_major_adverbs.count(a)))
arranged_major_adverbs.sort(key=lambda x: x[1], reverse=True)

for_major_nouns_unique = list(set(for_major_nouns))
arranged_major_nouns = []
for i in range(0,len(for_major_nouns_unique)-1):
    a = for_major_nouns_unique[i]
    arranged_major_nouns.append((a, for_major_nouns.count(a)))
arranged_major_nouns.sort(key=lambda x: x[1], reverse=True)

for_major_verbs_unique = list(set(for_major_verbs))
arranged_major_verbs = []
for i in range(0,len(for_major_verbs_unique)-1):
    a = for_major_verbs_unique[i]
    arranged_major_verbs.append((a, for_major_verbs.count(a)))
arranged_major_verbs.sort(key=lambda x: x[1], reverse=True)
```

```python
# libraries
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
 
# Build a dataframe with your connections
df = pd.DataFrame({'from':lemma_words+stem_words, 'to':stem_words+origin_words})
 
# And a data frame with characteristics for your nodes
carac = pd.DataFrame({'ID':unique_words, 'myvalue':group_colors})
 
# Build your graph
G=nx.from_pandas_edgelist(df, 'from', 'to', create_using=nx.Graph())
 
# The order of the node for networkX is the following order:
G.nodes()
# Thus, we cannot give directly the 'myvalue' column to netowrkX, we need to arrange the order!
 
# Here is the tricky part: I need to reorder carac to assign the good color to each node
carac= carac.set_index('ID')
carac=carac.reindex(G.nodes())
 
# And I need to transform my categorical column in a numerical value: group1->1, group2->2...
carac['myvalue']=pd.Categorical(carac['myvalue'])
carac['myvalue'].cat.codes
 
# Custom the nodes:
nx.draw(G, with_labels=True, node_color=carac['myvalue'].cat.codes, cmap=plt.cm.Set1, node_size=1, font_size=1)
```
